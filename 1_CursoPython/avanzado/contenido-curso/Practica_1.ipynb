### PRACTICA 1: Lectura y Almacenamiento de Archivos en Multiples Formatos
#### Lectura de json desde la web (citibikenyc)

```python
from urllib.request import urlopen
import json

# Descargar la informacion disponible en el sitio Web < City Bike NYC >
repuesta = urlopen("https://raw.githubusercontent.com/codificandobits/pandas-nivel-basico-data/main/stations.json") # Return txt
repuesta = repuesta.read()

json_dict = json.loads(repuesta) # from "str" to "dict"
```

```python
# Informacion basica que se encuentra en el key "stationBeanList", que es una lista de Dictionarios

n_estaciones = len(json_dict['stationBeanList'])
print(f'Cant. Estaciones: {n_estaciones}')
print(f'\nEj: Datos de una estacion:\n{json_dict['stationBeanList'][2]}')
```

```python
# Creacion del archivo TXT solo con los campos deseados para este proyecto (credo a partir del JSON dict)

with open('BikesNY.txt', 'w') as archivo:
    archivo.write('NOMBRE EST., BIC. DISPONIBLE, DIR.\n')

    for estacion in json_dict['stationBeanList']:
        nombre = estacion['stationName']
        nbicis = estacion['availableBikes']
        direcc = estacion['stAddress1']
        
        if nbicis != 0:
            archivo.write(f'{nombre}, {nbicis}, {direcc}\n')
    
```

```python
# Creando otro archivo BikesNY.txt, pero con un formato mas "Tabular" para su mejor lectura

formato_cols = '{0:<35} {1:>17} {2:>35}'

with open('BikesNY_2.txt', 'w') as archivo:
    
    archivo.write(formato_cols.format('NOMBRE EST.', 'BIC. DISPONIBLE', 'DIR.\n'))

    for estacion in json_dict['stationBeanList']:
        nombre = estacion['stationName']
        nbicis = estacion['availableBikes']
        direcc = estacion['stAddress1']

        if nbicis != 0:
            archivo.write(formato_cols.format(nombre, nbicis, direcc +'\n'))
    
```

```python
# Creando otro archivo BikesNY, pero .csv. Este ya viene en formato mas "Tabular" para su mejor lectura y poder abrir en Excel

import csv

with open('BikesNY.csv', 'w') as archivo:
    
    my_writer = csv.writer(archivo)
    my_writer.writerow(['NOMBRE EST.', 'BIC. DISPONIBLE', 'DIR.'])

    for estacion in json_dict['stationBeanList']:
        nombre = estacion['stationName']
        nbicis = estacion['availableBikes']
        direcc = estacion['stAddress1']
        
        if nbicis != 0:
            fila = [nombre, nbicis, direcc]
            my_writer.writerow(fila)
```
