### 1. Excepciones (try-except)

```python
try:
    print("Hola Mundo")
except:
    print("Hubo una excepcion")
```

### 2. Excepciones (else-finally)

```python
                # ---------------- else
try:
    num1 = int(input("Introduce el numerador: "))
    num2 = int(input("Introduce el denominador: "))
    result = num1/num2
    
except ValueError:
    print("Debe introducior solo cantidades numericas")
except ZeroDivisionError:
    print("El denominador no puede ser Cero")
else: 
    print(f"La divicion es: {result}")
    
```

```python
                # ---------------- finally (este bloque se ejecutarA siempre, (SI o SI))
try:
    num1 = int(input("Introduce el numerador: "))
    num2 = int(input("Introduce el denominador: "))
    result = num1/num2
    
except ValueError:
    print("Debe introducior solo cantidades numericas")
except ZeroDivisionError:
    print("El denominador no puede ser Cero")
else: 
    print(f"La divicion es: {result}")

finally:
    print("\nYo siempre me ejecuto (Si o SI)")
    
```

### 3. Manejo de archivos de textos (Escribiendo en un Archivo de Texto)

```python
# with: prepara el espacio en memoria y tambien cierra el objecto al terminar.
with open('my_archivo_texto.txt', mode='w') as archivo:   
    archivo.write("Linea 1\n")
    archivo.write("Linea 2\n")
    archivo.write("Linea 3")
    #1. si el archivo no existe, este se crearA autiomaticamente. ( -- open('./my_ruta/my_archivo_texto.txt', mode='w') -- )
    #2. Se puede especificar el lugar donde quiero que se cree el archivo.
    
```

```python
#LEYENDO EL TXT CON 'enumerate'

with open('my_archivo_texto.txt', mode='r') as archivo:
    for linea, registro in enumerate(archivo): # enumerate: sirve para retornar la linea del registro y el registro en si.
        print(f"Registro: {linea} {registro}")
```

```python
# Guardando cada registro del TXT en una lista, con lista 'comprehension'

with open('my_archivo_texto.txt', mode='r') as archivo:
    registroTXT = [linea.lower().rstrip() for linea in archivo] # quitE mayuscula y espacios en blancos con '.rstrip()'
    print(f"Registro: {registroTXT}")
```

```python
# LEYENDO EL ARCHIVO TXT A PARTIR DE UNA LINEA ESPECIFICA CON METODO 'open().seek()'

with open('my_archivo_texto.txt', mode='r') as archivo:
    
    archivo.seek(8)   #no leerA la primera fila o registro
    for linea, registro in enumerate(archivo):
        print(f"Registro: {linea} {registro}")
```

```python
# Modificar un archivo de Texto con metodo "open().writelines()" (open en modo "a")

lista_append = ["\nNueva linea 1","\nNueva linea 2","\nNueva linea 3"]

with open('my_archivo_texto.txt', mode='a') as archivo: # mode='a' permite editar una parte el texto, a diferencia de "w" que reescribe todo
    archivo.writelines(lista_append)

with open('my_archivo_texto.txt', mode='r') as archivo:
    for i in archivo:
        print(i)
```

```python
# Actulizacion de texto en un punto Intermedio (ActualizarE la linea 3)

with open('my_archivo_texto.txt', mode='r') as archivo:
    lineas = archivo.readlines()
    
    print(f'TXT original:\n {lineas}\n')
    
    for i, linea in enumerate(lineas):
        if linea.startswith('Linea 3'): # Buscando el patron que quiero modificar
            lineas[i]='Linea Actualizada 3'

    print(f'Como se verA despues:\n {lineas}\n')
    
```

```python
# Actulizacion de texto en un punto Intermedio (reescribiendo el txt por completo, pero solo cambiarA la liena 3)

with open('my_archivo_texto.txt', mode='w') as archivo: # En mode="w" para reescribir todo el txt con 'lineas'
    archivo.writelines(lineas)
```

### 3. Manejo de archivos de JSON

```python
# creando archivo JSON ( dump(<Src file>, <destine>) )

import json

dict_src ={'Empleado':['Carla', 'Miguel', 'Angel', 'Alturo',], 'Trabajo':['Ingeniero civil', 'Contable', 'Profesor', 'Psicologia']}
with open('EmpleadoJsn.json', mode='w') as archivo:
    json.dump(dict_src, archivo)
```

```python
#Leyendo un JSON y cargandolo en un DICT. ( load(<Src file>) )  ---> return: dict

with open('EmpleadoJsn.json', mode='r') as archivo:
    download_json = json.load(archivo)

print(download_json)
```

```python
# LEER un Dict que fue creado a partir de un JSON, de una forma mas amigable con "dumps()" e indent='3'

print(json.dumps(download_json, indent=2))
    
```
