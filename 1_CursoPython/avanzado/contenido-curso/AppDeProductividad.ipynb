```python
import textwrap

class Tarea():
    def __init__(self, descripcion, categoria, detalle):
        self._descripcion = descripcion
        self._categoria = categoria
        self._detalle = detalle
        self._estado = "Pendiente"

        # ================== GETTERS =====================
    @property
    def descripcion(self):
        return self._descripcion
            
    @property
    def categoria(self):
        return self._categoria
        
    @property
    def detalle(self):
        return self._detalle

    @property
    def estado(self):
        return self._estado

    # ================== SETTER =====================
    #@estado.setter
    def cambio_estado(self):
        if self._estado == 'Pendiente':
            self._estado = 'Completada'
        else:
            self._estado = 'Pendiente'
    
        
    # ==================== MOSTRAR ===================
    def mostrar(self):
        # SerA el ancho maximo del texto de 'detalle' (wrapping)
        ancho = 80

        #wrapping
        str_detalle = f'Detalle: self._detalle'
        if len(str_detalle) > ancho:
            str_detalle = textwrap.fill(str_detalle, width=ancho)
        else:
            pass

        print('*' * ancho)
        print(f"Tarea: {self._descripcion}")
        print(f"Categoria: {self._categoria}")
        print(f"Estado: {self._estado}\n")
```

```python
import os
import csv

class App:
    def __init__(self):
        
        """Crea una instancia de Administrador y crea o abre la BD csv si ya existe"""
        self.administrador = Administrador() #Instancia de administrador
        self._abrir_BD() #Abrir o crear la BD csv
        self._ejecutar() # Iniciar la APP
        
        
    def _abrir_BD(self):

        # Verificar si la BD existe antes de intentar abrirla, en caso de no existir, la crea en el else.
        if os.path.isfile('./tarea_BD.csv'): 
            with open('tarea_BD.csv', mode='r') as archivo:
                self._agregar_tarea()
                self.administrador.mostrar()  
        else:
            with open('./tarea_BD.csv', mode='w') as archivo:
                #print("bd")    
                pass
                
    def _agregar_tarea(self):
        #Lee desde la BD y escribe esos datos guardados en administrador para mostrarla"""
        """Leer desde la BD e iterar sobre cada fila, luego hacer una Tarea para cada fila de la BD, luego agregar esa tarea al Administrador"""
        
        with open('tarea_BD.csv', mode='r') as archivo:
            reader = csv.reader(archivo, delimiter=',', quotechar='"')
        
            for fila in reader:
                if len(fila) != 0: # Esta validacion la estoy haciendo, porque en mi BD se estan guardano filas (listas) vacias, cosa que hay que chequear
                    tarea = Tarea(fila[0], fila[1], fila[2])
                    #print(fila)
                    if fila[3] == "Completada":
                        tarea.estado = "Completada"
                    self.administrador.agregar_tarea(tarea)
                    #print('UNA TAREA AGREGADA')
                


    def _ejecutar(self):
        """Aplicativo para la interacion del Usuario via teclado"""

        continuar = True
        while continuar:
            print("\n\nSeleccione una opcion:")
            print("    (1) Mostrar administrador de tareas")
            print("    (2) Agregar tarea")
            print("    (3) Actualizar tarea")
            print("    (4) Eliminar tarea")
            print("    (5) Ver datalle de una tarea")
            print("    (6) Salir")

            opcion = int(input('Opcion: '))
            
            if opcion == 1:
                """Mostrar administrador de tarea"""
                self.administrador.mostrar(id_tarea=0) #Pendiente de ajustar la opcion para que el
                                                        #usuario pueda buscar una tarea especifica por id

            elif opcion == 2:
                """Agregar Tarea; Crea un obj Tarea y luego lo agrega a la clase administrador"""
                print("\n\t Agregar tarea:")
                descripcion = str(input('Descripcion: '))
                categoria = str(input('Categoria: '))
                detalle = str(input('Detalle: '))

                tarea = Tarea(descripcion, categoria, detalle)
                self.administrador.agregar_tarea(tarea) # El administrador muestra un mensaje indicando que agregO con exito
                #print("Tarea agregada con exito!!!\n")
                self.administrador.mostrar(id_tarea=0)
                
            elif opcion == 3:
                """Actualiza el "Estado" de la tarea en base a su Indice (ID)"""

                try:
                    upID = int(input("Ingrese el ID de la Tarea que quiere actualizar: "))
                    self.administrador.actualizar_tarea(upID)
                    print()
                except:
                    print("""\nOcurrio un Error al actualizar la Tarea, asegurese que el ID exista y su valor de entrada sea un numero""")

            
            elif opcion == 6:
                """Salir del menu y guardar cambios en la BD"""

                # AGUARDAR TODO EN LA BASE DE DATOS
                with open('./tarea_BD.csv', mode='w') as archivo:
                    """Abre la BD en modo w y se guarda cada nueva tarea que existe en el Administrador, en la BD tarea_BD.csv"""
                    
                    writer = csv.writer(archivo, delimiter=',') #, quotechar='"'
                    for tarea in self.administrador.tarea: #Se esta Iterando sobre una lista que contiene los objectos Tarea                        
                        writer.writerow([tarea.descripcion, tarea.categoria, tarea.detalle, 'Pendiente']) #GUARDA todo en la BD
                        #print(tarea.descripcion)
                        
                    print("Se GuardO todo correctamente!!!")
                
                continuar = False #Cerrar el bucle while (cerrar app)

                
                
            else:
                print("Ingrese un numero valido, entre 1 y 6")
            

                    
            

    
```
