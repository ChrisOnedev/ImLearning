```python
import textwrap

class Tarea():
    def __init__(self, descripcion, categoria, detalle):
        self._descripcion = descripcion
        self._categoria = categoria
        self._detalle = detalle
        self._estado = "Pendiente"

        # ================== GETTERS =====================
    @property
    def descripcion(self):
        return self._descripcion
            
    @property
    def categoria(self):
        return self._categoria
        
    @property
    def detalle(self):
        return self._detalle

    @property
    def estado(self):
        return self._estado

    # ================== SETTER =====================
    #@estado.setter
    def cambio_estado(self):
        if self._estado == 'Pendiente':
            self._estado = 'Completada'
        else:
            self._estado = 'Pendiente'
    
        
    # ==================== MOSTRAR ===================
    def mostrar(self):
        # SerA el ancho maximo del texto de 'detalle' (wrapping)
        ancho = 80

        #wrapping
        str_detalle = f'Detalle: self._detalle'
        if len(str_detalle) > ancho:
            str_detalle = textwrap.fill(str_detalle, width=ancho)
        else:
            pass

        print('*' * ancho)
        print(f"Tarea: {self._descripcion}")
        print(f"Categoria: {self._categoria}")
        print(f"Estado: {self._estado}\n")
```

### Implementacion de la clase Administrador

```python
class Administrador():
    def __init__(self):
        self._tarea = []

    @property
    def tarea(self):
        """Propiedad para poder acceder al atributo privado _tarea"""
        return self._tarea

    
    def mostrar(self, id_tarea=0):
        """Mostrar los detalles de las Tareas, puedo ver una sola Tarea o todas en formato tabla"""

        formato_cols = '{0:<10} {1:<20} {2:<20} {3:<20}' # id, tarea, categoria, estado
        print(formato_cols.format('ID', 'TAREA', 'CATEGORIA', 'ESTADO'))
        
        if id_tarea == 0:
            for id, tarea in enumerate(self._tarea):
                print(formato_cols.format(
                    id+1
                    ,tarea.descripcion
                    ,tarea.categoria
                    ,tarea.estado
                ))
        else:
            for id, tarea in enumerate(self._tarea):
                if id == id_tarea-1:
                    print(formato_cols.format(
                        id+1
                        ,tarea.descripcion
                        ,tarea.categoria
                        ,tarea.estado
                    ))    

    def agregar_tarea(self, tarea=None):
        """Agregar nueva tarea a la lista de tareas"""
        if tarea != None:
            self._tarea.append(tarea)
            print("Tarea agregada correctamente!!!")
        else:
            print('Ninguna Tarea fue cedida para agregarla durante el proceso de agregar nueva tarea')

    
    def actualizar_tarea(self, id_tarea):
        """Actualiza el "Estado" de la tarea en base a su Indice (ID)"""
        if len(self._tarea) !=0:
            
            self._tarea[id_tarea-1].cambio_estado()
            print(f"Es estado de la Tarea N° {id_tarea} fue actualizada a {self._tarea[id_tarea-1].estado}")
            
        else:
            print("La lista de Tarea esta vacia")
            #print(f"La Tarea N° {id} no existe en la lista de Tarea")


    def eliminar_tarea(self):
        pass
        

    
```
